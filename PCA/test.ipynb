```python
# load dataset
import pandas as pd
mnist = pd.read_csv(r'C:\Users\ASUS TUF\projects\ML3\mnist.csv', header=None)
# split dataset into labels and pixels
labels = mnist.iloc[:, 0].values
pixels = mnist.iloc[:, 1:].values
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2).fit_transform(pixels)
import numpy as np
cov = np.cov(pca, rowvar=False)
print(cov)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10, max_iter=20, random_state=0)
kmeans.fit(pixels)
cluster_labels = kmeans.labels_
print(cluster_labels)
```

```python
from sklearn.cluster import KMeans
k_values = range(1, 11)
loss = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(pca)
    loss.append(kmeans.inertia_)

# Plot the loss curve
import matplotlib.pyplot as plt
plt.plot(k_values, loss, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Loss')
plt.title('K-Means Loss Curve')
plt.show()
```
